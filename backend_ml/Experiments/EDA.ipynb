```python
chars_to_delete = punctuation + "©\xa0\xad\t\n\r\x0b\x0c«»‘’£"
trans_table = str.maketrans('', '', chars_to_delete)
trans_table[ord('-')] = ' '


def lemmatize_text(text):
    tokens = mystem.lemmatize(text.lower())
    tokens = [token for token in tokens if token not in russian_stopwords\
              and token != " " \
              and token.strip() not in punctuation]
    
    text = " ".join(tokens)
    
    return text


def filter_text(text):
    text = text.replace('\\n', '').translate(trans_table)
    
    return re.sub(' +', ' ', text).lower()


def position_preprocessing(text):
    text = text.lower()
    result = []
    
    text = text.replace('.', ' ')
    texts = re.split('[,|\|/]', text)
    
    for text in texts:
        text = text.replace(';', ' ')
        text = text.replace('-', ' ')
        text = text.replace('/', ' ')
        text = re.sub(' +', ' ', text)

        text = text.replace(' зам ', ' заместитель ')
        text = text.replace(' зав ', ' заведующий ')
        text = text.replace(' стар ', ' старший ')
        text = text.replace(' ст ', ' старший ')
        text = text.replace(' глав ', ' главный ')
        text = text.replace(' ген ', ' генеральный ')
        text = text.replace(' пом ', ' помощник ')
        text = text.replace(' нач ', ' начальник ')
        text = text.replace(' вод ', ' водитель ')
        text = text.replace(' мл ', ' младший ')
        text = text.replace(' дир ', ' директор ')
        text = re.sub(' +', ' ', text).strip()
        
        result.append(filter_text(text))
    
    return result
```

**Final processing**

```python
from resume_parser_ml import parser


def process_waiter(row):
    position, scopes = parser.get_postition_scopes(row['position'])
    exp = parser.get_experience(row['start'],
                                row['end'])
    menu_knowledge = parser.get_menu_knowledge(row['description'])
    vip_clients = parser.get_vip_clients(row['description'])
    is_cashier = parser.is_cashier(row['description'])
    cashier_type = parser.get_cashiers(row['description'])
    is_catering = parser.is_catering(row['description'])
    soft_skills = parser.get_skills(row['description'])
    company = row['organization']

    response = {'data': [["Моя позиция в организации:", position],
                        ['Отрасли, в которых я работаю:', ', '.join(scopes)],
                        ['Опыт работы:', str(exp) + ' лет'],
                        ['Работал в:', company],
                        ['Знания меню:', 'имеются' if menu_knowledge else ''],
                        ['Опыт работы с VIP клиентами:', 'есть' if vip_clients else ''],
                        ['Работал с кассой:', 'да' if is_cashier else'нет'],
                        ['Системы для работы с кассой:', ', '.join(cashier_type) if cashier_type else ''],
                        ['Услуги кейтеринга:', 'оказывал' if is_catering else ''],
                        ['Мои личные качества:', ', '.join(soft_skills)]]}
    return response


def process_driver(row):
    position, scopes = parser.get_postition_scopes(row['position'])
    exp = parser.get_experience(row['start'],
                                row['end'])

    category = parser.get_category(row['description'])
    auto_height = parser.get_auto_height(row['description'])
    in_out = parser.get_in_out(row['description'])
    auto_weight = parser.get_auto_weight(row['description'])
    shtabeler = parser.get_shtabeler(row['description'])
    company = row['organization']

    response = {'data': [["Моя позиция в организации:", position],
                        ['Отрасли, в которых я работаю:', ', '.join(scopes)],
                        ['Опыт работы:', str(exp) + ' лет'],
                        ['Работал в:', company],
                        ['Вожу автомобили категорий:', ', '.join(category)],
                        ['Автомобили в метрах не выше:', auto_height],
                        ['Оказываю услуги погрузки/выгрузки:', 'да' if in_out else ''],
                        ['Максимальная гружоность (тонны):', auto_weight],
                        ['Умею пользоваться штабелером:', 'да' if shtabeler else '']]}
    return response


def process_cook(row):
    position, scopes = parser.get_postition_scopes(row['position'])
    exp = parser.get_experience(row['start'],
                                row['end'])

    manufacture = parser.get_manufacture(row['description'])
    cuisuins = parser.get_cuisuin(row['description'])
    max_cap = parser.get_max_clients(row['description'])
    menu = parser.get_if_menu_created(row['description'])
    banquet = parser.get_if_banquet_holded(row['description'])
    inventory = parser.get_if_inventory_holded(row['description'])

    company = row['organization']

    response = {'data': [["Моя позиция в организации:", position],
                        ['Отрасли, в которых я работаю:', ', '.join(scopes)],
                        ['Опыт работы:', str(exp) + ' лет'],
                        ['Работал в:', company],
                        ['Умею готовить кухни:', ', '.join(cuisuins)],
                        ['Максимальная посадка, которую я обслуживал:', max_cap],
                        ['У меня был опыт создания меню:', 'да' if max_cap else ''],
                        ['Был опыт работы для банкетов:', 'да' if banquet else ''],
                        ['Умею производить инвенторизацию:', 'да' if inventory else '']]}
    return response


def process_undetected(row):
    position, scopes = parser.get_postition_scopes(row['position'])
    exp = parser.get_experience(row['start'],
                            row['end'])

    company = row['organization']

    response = {'data': [["Моя позиция в организации:", position],
                        ['Отрасли, в которых я работаю:', ', '.join(scopes)],
                        ['Опыт работы:', str(exp) + ' лет'],
                        ['Работал в:', company],
                        ['Моя специальность:', ''],
                        ['Мои личные качества:', ''],
                        ['Программы с которыми я работал:', '']]}
    return response


def process_data(row):
    resume_class = parser.get_position(row['position'])
    resume_class = parser.classes[resume_class]
    resume_class = parser.class_to_title[resume_class]

    if resume_class == 'Сфера обслуживания':
        return process_waiter(row)
    elif resume_class == 'Экспедитор':
        return process_driver(row)
    elif resume_class == 'Повар':
        return process_cook(row)
    elif resume_class == 'unknown':
        return process_undetected(row)

def get_no_experience():
    response = jsonify({'data': [["Моя позиция в организации:", ''],
                        ['Отрасли, в которых я работаю:', ''],
                        ['Опыт работы:', ''],
                        ['Работал в:', ''],
                        ['Моя специальность:', ''],
                        ['Мои личные качества:', ''],
                        ['Программы с которыми я работал:', '']]})
    return response


def proceed_data(row):
    response = process_data(row)
    return response

def no_experience():
    response = get_no_experience()
    return response
```
